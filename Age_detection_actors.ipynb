```python
# Read the image files.

import os
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

current_directory = os.getcwd()
current_train_directory_full_path = os.path.join(current_directory,'IMFDB/train_images/Train')
current_test_directory_full_path = os.path.join(current_directory,'IMFDB/test_images/Test')
train_labels_path = os.path.join(current_directory,'IMFDB/train_images/train.csv')
test_labels_path = os.path.join(current_directory, 'IMFDB/test_images/test.csv')
```

```python
train_labels = pd.read_csv(train_labels_path)
test_labels = pd.read_csv(test_labels_path)
```

```python
train_labels_list = train_labels['ID'].tolist()
test_labels_list = test_labels['ID'].tolist()

target_classes = train_labels['Class']
target_classes_df = pd.DataFrame(target_classes)
```

```python
train_labels_list_v1 = [os.path.join(current_train_directory_full_path, img_id) for img_id in train_labels_list]
test_labels_list_v1 = [os.path.join(current_test_directory_full_path, img_id) for img_id in test_labels_list]

train_labels_v1_df = pd.DataFrame(train_labels_list_v1, columns=['URL'])
```

```python
# Reading images as grayscale and saving them as numpy arrays.

train_ndarray_list = []
for num, img in enumerate(train_labels_list_v1):
    gray_image = cv2.imread(img)
    gray_img = cv2.cvtColor(gray_image,cv2.COLOR_BGR2GRAY)
    print("Shape", gray_img.shape)
    train_ndarray_list.append(gray_img)
       
```

```python
# Explore the Image Data set.Exploratory Image Analysis.


# Basic Exploration
# Check dimension and confirm its not 3-band (Grayscale Image) # Taking 15 images for test

for num,img in enumerate(train_ndarray_list):
    if num > 15:
        break
    else:
        print("Dimension :",np.shape(img))
        print("Min :", np.min(img) , "Max :", np.max(img))
        
        
```

```python
# Using Exploratory Data Analysis technique to discover patterns.
sns.catplot(x='Class', kind='count', data=target_classes_df, height=4, aspect=11);
```

```python
# Resize Image with constant aspect ratio and padding using OpenCV

def image_resize_copymake(imag, desired_size=128):
    im = cv2.imread(imag, 0)
    old_size = im.shape[:2]
    ratio = float(desired_size)/max(old_size)
    new_size = tuple([int(x*ratio) for x in old_size])
    
    # new_size should be in (width, height) format

    im = cv2.resize(im, (new_size[1], new_size[0]))
    delta_w = desired_size - new_size[1]
    delta_h = desired_size - new_size[0]
    top, bottom = delta_h//2, delta_h-(delta_h//2)
    left, right = delta_w//2, delta_w-(delta_w//2)
    color = [0, 0, 0]
    new_im = cv2.copyMakeBorder(im, top, bottom, left, right, cv2.BORDER_CONSTANT,value=color)
    print("New_im Shape", new_im.shape)

resized_image_list = []    
for img in train_labels_list_v1:
    resized_image = image_resize_copymake(img, 140)
    print(resized_image)
    resized_image_list.append(resized_image)
```
